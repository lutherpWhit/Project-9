```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
Works = pd.read_csv("works-20210226.csv")
Tags = pd.read_csv("tags-20210226.csv")
```

```python
Works["datetime"] = pd.to_datetime(Works["creation date"])
Works["datetime"].head()
```

```python
Works["tag list"] = Works["tags"].str.split('+')
```

```python
Fandoms = Tags[Tags["type"] == "Fandom"]
Fandoms.head()
```

```python
idList = Fandoms.sort_values(by = ["cached_count"], ascending = False).head(n=10)["id"].to_list()
idList
```

```python
idList_str = []
for item in idList:
    idList_str.append(str(item))
idList_str
```

```python
Works
```

Here I will create the function to find if each of the top ten fandoms are in the tags list of any given work

```python
idList
```

```python
def tagFind(tagList):
    fandoms = []
    for item in tagList:
        for tag in idList_str:
            if tag == item:
                fandoms.append(tag)
            
    if len(fandoms) != 0:
        return fandoms
    else:
        return False
```

testing the function on just a month to be sure

```python
Works_2020 = Works[Works["datetime"].dt.year == 2020]
```

```python
January = Works_2020[Works_2020["datetime"].dt.month == 1]
January.head()
```

```python
January["tag list"].apply(tagFind).value_counts() #seems to work
```

```python
January["tag list"].apply(tagFind)
```

```python
January["fandoms"] = January["tag list"].apply(tagFind)
January["fandoms"].value_counts()
```

```python
January[January["fandoms"] != False]
```
